```python
import numpy as np
from librosa.core import istft, load, stft, magphase
import soundfile as sf
from keras.models import load_model
from config import *
```

```python
AUDIO_PATH = '/content/drive/MyDrive/modelv_5/predict/mix.wav'
```

```python
model = load_model('/content/drive/MyDrive/modelv_5/model/piano.h5')
```

```python
mix_wav, _ = load(AUDIO_PATH, sr=44000)
mix_wav_mag, mix_wav_phase = magphase(stft(mix_wav, n_fft=window_size, hop_length=hop_length))

START = 0
END = START + 128

mix_wav_mag=mix_wav_mag[:, START:END]
mix_wav_phase=mix_wav_phase[:, START:END]

X=mix_wav_mag[1:].reshape(1, 512, 128, 1)

y=model.predict(X, batch_size=32)

target_pred_mag = np.vstack((np.zeros((128)), y.reshape(512, 128)))
```

```python
y = istft((target_pred_mag * mix_wav_phase), win_length=window_size, hop_length=hop_length)
```

```python
sf.write('/content/drive/MyDrive/modelv_5/predict/target.wav', y , 44000)
```
